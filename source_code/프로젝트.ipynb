```python
df = pd.read_csv('../data/서울특별시 부동산 실거래가 정보.csv', encoding="UTF-8")
df.head()
```

```python
df.columns
```

```python
view_columns = ['자치구명','신고년도','건물주용도','물건금액','건물면적']
df_1 = df[view_columns]
df_1.head()
```

```python
df_apt=df_1[(df_1['건물주용도']=='아파트') & (df['신고년도']>2016)]
df_apt.head()
```

```python
# 소수점 2번째 자리까지 표기
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_apt['면적당 금액'] = (df_apt['물건금액']/df_apt['건물면적'])
df_apt.head()
```

```python
df_apt['자치구명'].unique()
```

```python
df_result = df_apt[['자치구명','면적당 금액']]
df_result.head(10)
```

```python
df_gu =df_result.unstack(level=0)
df_gu
```

```python
df_gu_group = df_result.groupby(['자치구명'], as_index=False).mean()
df_gu_group
```
