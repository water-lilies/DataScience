```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline 

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
df = pd.read_csv('../data/서울특별시 부동산 실거래가 정보.csv', encoding="UTF-8")

# 컬럼명으로 데이터 추출
view_columns = ['자치구명','신고년도','건물주용도','물건금액','건물면적']
df_1 = df[view_columns]

# 조건에 맞는 행 추출
df_apt=df_1[(df_1['건물주용도']=='아파트') & (df['신고년도']>2016)]

# 소수점 2번째 자리까지 표기
pd.options.display.float_format = '{:.2f}'.format

# 컬럼 추가
df_apt['면적당 금액'] = (df_apt['물건금액']/df_apt['건물면적'])

df_gu = df_apt[['자치구명','면적당 금액']]

# 자치구명 단위로 묶은 평균값
df_gu_group = df_gu.groupby(['자치구명'], as_index=False).mean()
df_gu_group
```

```python
df_pop = pd.read_excel('../data/서울시 주민등록인구 (구별) 통계.xls',encoding="UTF-8")
df_pop.head()
```

```python
df_fore = df_pop[['자치구','인구','인구.6']]
df_fore.head()
```

```python
df_foreign = df_fore.iloc[3:,:]
df_foreign.columns = ['자치구명','인구 총 합계', '외국인 인구수']
df_foreign['외국인 비율'] = (df_foreign['외국인 인구수']/df_foreign['인구 총 합계'] * 100)
df_foreign
```

```python
df_seni = df_pop[['자치구','인구','65세이상고령자']]
df_senior = df_seni.iloc[3:,:]
df_senior.columns = ['자치구명','인구 총 합계', '65세이상고령자']
df_senior['고령자 비율'] = (df_senior['65세이상고령자']/df_senior['인구 총 합계'] * 100)
df_senior
```

```python
df_1 = pd.merge(df_gu_group, df_foreign, on='자치구명')
df_2 = pd.merge(df_1, df_senior, on = '자치구명')
df_2
```

```python
df_result = df_2[['자치구명','면적당 금액','외국인 비율','고령자 비율']]
df_result
```
